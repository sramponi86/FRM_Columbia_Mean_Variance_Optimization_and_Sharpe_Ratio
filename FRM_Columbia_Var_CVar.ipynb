```python
import numpy as np
import pandas as pd

# Load your historical return data (e.g., CSV with daily returns for multiple stocks)
data = pd.read_excel('Test.xlsx', index_col='date')  #, on_bad_lines='skip')
data
```

```python
# Compute daily losses
daily_losses = - data
daily_losses
```

```python
# Initialize dictionaries to store risk measures for each stock
variance_dict = {}
var_0_9_dict = {}
cvar_0_9_dict = {}

# Calculate risk measures for each stock
for stock in daily_losses.columns:
    # Calculate variance
    variance = daily_losses[stock].var()
    variance_dict[stock] = variance

    # Calculate VaR at 0.9 level
    var_0_9 = daily_losses[stock].quantile(0.9)
    var_0_9_dict[stock] = var_0_9

    # Calculate CVaR at 0.9 level
    cvar_0_9 = daily_losses[stock][daily_losses[stock] >= var_0_9].mean()
    cvar_0_9_dict[stock] = cvar_0_9

# Find stocks with the highest values for each risk measure
highest_variance_stock = max(variance_dict, key=variance_dict.get)
highest_var_stock = max(var_0_9_dict, key=var_0_9_dict.get)
highest_cvar_stock = max(cvar_0_9_dict, key=cvar_0_9_dict.get)

# Print results
print(f"Stock with highest Variance: {highest_variance_stock}, Variance: {variance_dict[highest_variance_stock]:.4f}")
print(f"Stock with highest VaR 0.9: {highest_var_stock}, VaR 0.9: {var_0_9_dict[highest_var_stock]:.4f}")
print(f"Stock with highest CVaR 0.9: {highest_cvar_stock}, CVaR 0.9: {cvar_0_9_dict[highest_cvar_stock]:.4f}")
```

```python
daily_portfolio_returns = data.mean(axis=1)

# Compute daily portfolio losses
daily_portfolio_losses = - daily_portfolio_returns

# Calculate VaR at 0.9 level (90th percentile of daily portfolio losses)
var_0_9_portfolio = daily_portfolio_losses.quantile(0.9)

# Print the result rounded to four decimal places
print(f"{var_0_9_portfolio}")
```

```python
# Calculate CVaR at 0.9 level (average of losses exceeding VaR 0.9)
cvar_0_9_portfolio = daily_portfolio_losses[daily_portfolio_losses >= var_0_9_portfolio].mean()

# Print the result rounded to four decimal places
print(f"{cvar_0_9_portfolio}")
```
