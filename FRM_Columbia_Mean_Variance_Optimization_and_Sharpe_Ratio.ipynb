```python
import numpy as np
from scipy.optimize import minimize

def portfolio_return(weights, mean_returns):
    return np.dot(weights, mean_returns)

def portfolio_volatility(weights, cov_matrix):
    return np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
```

```python
def minimize_portfolio_risk(weights, cov_matrix):
    return portfolio_volatility(weights, cov_matrix)
```

```python
constraints = [
    {'type': 'eq', 'fun': lambda x: np.sum(x) - 1},  # Sum of weights = 1
    {'type': 'ineq', 'fun': lambda x: portfolio_return(x, mean_returns) - 0.0005}  # Return >= 0.0005
]
```

```python
bounds = [(0, 1) for _ in range(len(mean_returns))]
```

```python
initial_weights = np.array([1/len(mean_returns)] * len(mean_returns))

# Optimize
optimal_portfolio = minimize(minimize_portfolio_risk, initial_weights, args=(cov_matrix),
                             method='SLSQP', bounds=bounds, constraints=constraints)

optimal_variance = optimal_portfolio.fun
print("Optimal portfolio variance (risk):", optimal_variance)
```

```python
optimal_weights = optimal_portfolio.x
expected_return_x1 = np.dot(optimal_weights, mean_returns)
round(expected_return_x1, 4)
```

```python
constraints_new = [
    {'type': 'eq', 'fun': lambda x: np.sum(x) - 1},  # Sum of weights = 1
    {'type': 'ineq', 'fun': lambda x: portfolio_return(x, mean_returns) - 0.0008}  # Return >= 0.0008
]

optimal_portfolio_new = minimize(minimize_portfolio_risk, initial_weights, args=(cov_matrix),
                                 method='SLSQP', bounds=bounds, constraints=constraints_new)

round(optimal_portfolio_new.fun, 6)
```

```python
constraints_x3 = [
    {'type': 'eq', 'fun': lambda x: np.sum(x) - 1},  # Sum of weights = 1
    {'type': 'ineq', 'fun': lambda x: portfolio_return(x, mean_returns) - 0.001}  # Return >= 0.0008
]

optimal_portfolio_x3 = minimize(minimize_portfolio_risk, initial_weights, args=(cov_matrix),
                                 method='SLSQP', bounds=bounds, constraints=constraints_x3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

phi_values = np.linspace(-2, 2, 100)
portfolio_returns = []
portfolio_variances = []

for phi in phi_values:
    # Calculate portfolio weights for given φ
    z = phi * optimal_portfolio.x + (1 - phi) * optimal_portfolio_new.x

    # Calculate portfolio return and variance
    portfolio_return = np.dot(z, mean_returns)
    portfolio_variance = np.dot(z.T, np.dot(cov_matrix, z))

    # Store the results
    portfolio_returns.append(portfolio_return)
    portfolio_variances.append(portfolio_variance)

plt.plot(portfolio_variances, portfolio_returns, label='Efficient Frontier')
plt.xlabel('Variance')
plt.ylabel('Expected Return')
plt.title('Efficient Frontier for Portfolios Generated by Two-Fund Theorem')
plt.legend()
plt.show()
```

```python
filtered_returns = [ret for ret, var in zip(portfolio_returns, portfolio_variances) if var <= 0.000040]

# Get the maximum return meeting the variance constraint
optimal_return = max(filtered_returns)
optimal_return_rounded = round(optimal_return, 4)
print("Approximate optimal return with variance <= 0.000040:", optimal_return_rounded)
```

```python
risk_free_rate = 0.0001

excess_returns = mean_returns - risk_free_rate

def minimize_portfolio_risk(x, cov_matrix):
    return np.dot(x.T, np.dot(cov_matrix, x))

constraints = [
    {'type': 'eq', 'fun': lambda x: np.dot(excess_returns, x) - 1},  # Excess return constraint
]

bounds = [(None, None) for _ in range(len(mean_returns))]

initial_weights = np.ones(len(mean_returns))

result = minimize(minimize_portfolio_risk, initial_weights, args=(cov_matrix),
                  method='SLSQP', bounds=bounds, constraints=constraints)

x_sharpe = result.x / np.sum(result.x)
portfolio_return = np.dot(x_sharpe, mean_returns)

portfolio_volatility = np.sqrt(np.dot(x_sharpe.T, np.dot(cov_matrix, x_sharpe)))

sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_volatility
sharpe_ratio_rounded = round(sharpe_ratio, 3)

print("Sharpe Ratio:", sharpe_ratio_rounded)
```
